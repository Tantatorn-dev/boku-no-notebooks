# Data Preparation
Let's import a dataset and preprocess it.

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```
df = pd.read_csv("/content/drive/My Drive/dataset/esl.csv")
df
```

```
# Let's drop some columns
df = df.drop(columns=["Timestamp","Z-H","คะแนน","น้องสามารถอดนอนเป็นเวลา 2 วัน 1 คืนได้ไหม"])
```

```
# let's drop test rows
df = df.drop(index=[56,57])
```

```
df
```

# Exploratory Data Analysis
Let's explore our data

```
n_men = len(df[df['คำนำหน้าชื่อ'] == "นาย"])
n_women = len(df[df['คำนำหน้าชื่อ'] == "นางสาว"])

plt.pie([n_men,n_women],labels=["male","female"],autopct='%1.1f%%')
plt.title('Ratio between male and female')
plt.axis('equal')
plt.show()
```

```
n_sec_17 = len(df[df['Sec ที่เรียน '] == 17])
n_sec_18 = len(df[df['Sec ที่เรียน '] ==  18])
n_sec_19 = len(df[df['Sec ที่เรียน '] ==  19])

plt.pie([n_sec_17,n_sec_18,n_sec_19],labels=["sec 17","sec 18","sec 19"],autopct='%1.1f%%')
plt.title('Section')
plt.axis('equal')
plt.show()
```
