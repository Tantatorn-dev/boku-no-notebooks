# Exploratory Data Analysis
Let's explore our data

```
# change font color
import matplotlib as mpl

COLOR = 'white'
mpl.rcParams['text.color'] = COLOR
mpl.rcParams['axes.labelcolor'] = COLOR
mpl.rcParams['xtick.color'] = COLOR
mpl.rcParams['ytick.color'] = COLOR
```

```
n_men = len(df[df['คำนำหน้าชื่อ'] == "นาย"])
n_women = len(df[df['คำนำหน้าชื่อ'] == "นางสาว"])

plt.pie([n_men,n_women],labels=["male","female"],autopct='%1.1f%%')
plt.title('Ratio between male and female')
plt.axis('equal')
plt.show()
```

```
n_sec_17 = len(df[df['Sec ที่เรียน '] == 17])
n_sec_18 = len(df[df['Sec ที่เรียน '] ==  18])
n_sec_19 = len(df[df['Sec ที่เรียน '] ==  19])

plt.pie([n_sec_17,n_sec_18,n_sec_19],labels=["sec 17","sec 18","sec 19"],autopct='%1.1f%%')
plt.title('Section')
plt.axis('equal')
plt.show()
```

```
interesting_topic = []

for responses in df['ตอนนี้น้องๆกำลังสนใจเทคโนโลยีใดเป็นพิเศษ']:
  for i in responses.split(", "):
    interesting_topic.append(i)
  
print(interesting_topic)  
```

```
from collections import Counter

topic = list(Counter(interesting_topic).keys()) 
topic_count = list(Counter(interesting_topic).values())
print(topic)
print(topic_count)
```

```
plt.pie(topic_count, labels=topic, autopct='%1.1f%%')
plt.title('Interesing topic')
plt.axis('equal')
plt.show()
```

## Word Cloud
create word clouds by participants' responses.

```
regexp = r"[ก-๙a-zA-Z']+"

path = '/content/drive/My Drive/dataset/charmonman.ttf'

soft_words = ['ที่','กัน','ก็','เป็น','อยู่','คือ','ครับ','ค่ะ','คะ','นะ','ด้วย','จะ','เพราะ','แต่','กับ']

wordcloud = WordCloud(
                      font_path=path, 
                      relative_scaling = 0.3,
                      min_font_size=1,
                      background_color="white",
                      width=1024,
                      height=768,
                      max_words=2000, 
                      colormap='plasma', 
                      scale=3,
                      font_step=4,
                      collocations=False,
                      regexp=regexp,
                      margin=2
                      )
```

```
from deepcut import tokenize

words = []
responses = df["อยากร่วมงานกับเพื่อนร่วมทีมแบบไหน"]

for response in responses:
  for word in deepcut.tokenize(response):
    words.append(word)

words = [x for x in words if x not in soft_words]

text = " ".join(words)
wordcloud.generate(text)

fig, ax = plt.subplots(1, 1, figsize=(16, 12))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
fig.show()
```
