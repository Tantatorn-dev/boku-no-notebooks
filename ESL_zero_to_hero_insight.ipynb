```
interesting_topic = []

for responses in df['‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ‡∏ô‡πâ‡∏≠‡∏á‡πÜ‡∏Å‡∏≥‡∏•‡∏±‡∏á‡∏™‡∏ô‡πÉ‡∏à‡πÄ‡∏ó‡∏Ñ‡πÇ‡∏ô‡πÇ‡∏•‡∏¢‡∏µ‡πÉ‡∏î‡πÄ‡∏õ‡πá‡∏ô‡∏û‡∏¥‡πÄ‡∏®‡∏©']:
  for i in responses.split(", "):
    interesting_topic.append(i)
  
print(interesting_topic)  
```

```
from collections import Counter

topic = list(Counter(interesting_topic).keys()) 
topic_count = list(Counter(interesting_topic).values())
print(topic)
print(topic_count)
```

```
plt.pie(topic_count, labels=topic, autopct='%1.1f%%')
plt.title('Interesing topic')
plt.axis('equal')
plt.show()
```

## Word Cloud
create word clouds by participants' responses.

```
%pip install deepcut
```

```
regexp = r"[‡∏Å-‡πôa-zA-Z']+"

path = '/content/drive/My Drive/dataset/charmonman.ttf'

wordcloud = WordCloud(
                      font_path=path, 
                      relative_scaling = 0.3,
                      min_font_size=1,
                      background_color="white",
                      width=1024,
                      height=768,
                      max_words=2000, 
                      colormap='plasma', 
                      scale=3,
                      font_step=4,
                      collocations=False,
                      regexp=regexp,
                      margin=2
                      )
```

```
from deepcut import tokenize

words = []
responses = df["‡∏≠‡∏¢‡∏≤‡∏Å‡∏£‡πà‡∏ß‡∏°‡∏á‡∏≤‡∏ô‡∏Å‡∏±‡∏ö‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏ô‡∏£‡πà‡∏ß‡∏°‡∏ó‡∏µ‡∏°‡πÅ‡∏ö‡∏ö‡πÑ‡∏´‡∏ô"]

for response in responses:
  for word in deepcut.tokenize(response):
    words.append(word)

text = " ".join(words)
wordcloud.generate(text)

fig, ax = plt.subplots(1, 1, figsize=(16, 12))
ax.imshow(wordcloud, interpolation='bilinear')
ax.axis("off")
fig.show()
```
